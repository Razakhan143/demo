```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.arange(0,10)
y=np.arange(5,15)
```

Line Plot

```python
plt.figure(figsize=((5,5)),dpi=100)
plt.title("practice graph")
plt.plot(x,y**1.9,label="1st",color="green",linewidth=2,marker='x',linestyle="--",markersize=10, markeredgecolor="red")
plt.plot(x[:6],y[:6]**2,'b',label="2nd")
plt.plot(x[5:],y[5:]**2,'b--')
plt.xticks((x**2)/8)
plt.yticks((y**2)*1.1)
plt.xlabel("price")
plt.ylabel("year")
plt.legend()
plt.show()
```

BAR PLOT

```python
# Data
labels = ["Data Science", "Web Development", "Software Engineering", "AI Engineering"]
ratings = [9, 7, 7, 8]

# Create the bar chart
plt.figure(figsize=(7, 5))  # Adjust figure size
plt.bar(labels, ratings, color=['skyblue', 'orange', 'green', 'purple'], alpha=0.8, edgecolor='black')

# Add titles and labels
plt.title("Field Ratings Over the Years", fontsize=14, fontweight='bold')
plt.xlabel("Fields", fontsize=12)
plt.ylabel("Ratings", fontsize=12)

plt.yticks(np.arange(0,11))
# Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the chart
plt.tight_layout()  # Adjust layout to avoid clipping
plt.show()
```

Scatter plot

```python
#Data
x=np.random.randint(0,101,size=(50))
y=np.random.randint(0,101,size=(50))
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(x, y, color='green', marker='x', alpha=0.6, label="Data Points")

plt.xlabel("X Data", fontsize=12)
plt.ylabel("Y Data", fontsize=12)
plt.title("Scatter Plot of X vs Y", fontsize=14, fontweight='bold')
plt.legend(loc="upper left")

plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

Histogram plot

```python
labels = [["Reading"]*8, ["Traveling"]*9, ["Gaming"]*7, ["Cooking"]*6]
# Create the bar chart
plt.figure(figsize=(7, 5)) 

plt.hist(labels,bins=1,color=['olive', 'tan', 'saddlebrown', 'forestgreen'], alpha=0.8, edgecolor='black',rwidth=100,align="mid")

# Add titles and labels
plt.title("Hobbies Ratings Over the Years", fontsize=14, fontweight='bold')
plt.xlabel("hobbies", fontsize=12)
plt.ylabel("Ratings", fontsize=12)

plt.yticks(np.arange(0,11))
# Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the chart
plt.tight_layout()  # Adjust layout to avoid clipping
plt.show()
```
