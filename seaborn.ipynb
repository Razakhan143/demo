```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import pandas as pd
data=sns.load_dataset('tips')
```

```python
data.head()
```

Relational / Statical Plot

Scatter plot

```python
sns.set_style("whitegrid")

sns.relplot(
    x="total_bill", y="tip", hue="time", style="sex", data=data,
     height=4, aspect=1.5
)

plt.title("Relationship Between Total Bill and Tip", fontsize=16, fontweight='bold')
plt.xlabel("Total Bill (in USD)", fontsize=12)
plt.ylabel("Tip (in USD)", fontsize=12)


plt.show()
```

Plotting With Numerical and categorical data

Strip plot / Swarm plot

```python
#strip plot
sns.catplot(
    x="day", y="tip", hue="day", kind="strip", data=data, 
    jitter=0.15, height=4, aspect=1.5, s=15
)

plt.title("Tip Distribution Across Days", fontsize=16, fontweight='bold')
plt.xlabel("Day of the Week", fontsize=12)
plt.ylabel("Tip Amount (in USD)", fontsize=12)

plt.show()
```

```python
#swarm plot
sns.catplot(
    x="day", y="tip", hue="sex", kind="swarm", data=data, 
     height=4, aspect=1.5, s=15 
)

plt.title("Tip Distribution Across Days", fontsize=16, fontweight='bold')
plt.xlabel("Day of the Week", fontsize=12)
plt.ylabel("Tip Amount (in USD)", fontsize=12)

plt.show()
```

Box Plot

can be use for both

```python
sns.catplot(
    x="smoker", y="tip", hue="smoker", data=data, kind="box", 
     height=4, aspect=1
)

# Add title and labels
plt.title("Tip Distribution by Day and Gender", fontsize=16, fontweight='bold')
plt.xlabel("Day of the Week", fontsize=12)
plt.ylabel("Tip Amount (in USD)", fontsize=12)

# Display the plot
plt.show()
```

```python
sns.catplot(
    x="day", y="tip", hue="sex", data=data, kind="box", 
     height=4, aspect=1.5
)

# Add title and labels
plt.title("Tip Distribution by Day and Gender", fontsize=16, fontweight='bold')
plt.xlabel("Day of the Week", fontsize=12)
plt.ylabel("Tip Amount (in USD)", fontsize=12)

# Display the plot
plt.show()
```

Violin plot

combination of box and kde plot


```python
# plt.style.use("fivethirtyeight")
```

```python
plt.figure(figsize=(6, 4))
sns.violinplot(
    x='day', y='total_bill', hue="smoker", data=data, 
)

plt.title("Total Bill Distribution by Day and Smoking Status", fontsize=16, fontweight='bold')
plt.xlabel("Day of the Week", fontsize=12)
plt.ylabel("Total Bill (in USD)", fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.catplot(
    x='day', y='total_bill', hue="sex", data=data,kind="violin",split=True
)

plt.title("Total Bill Distribution by Day and Smoking Status", fontsize=16, fontweight='bold')
plt.xlabel("Day of the Week", fontsize=12)
plt.ylabel("Total Bill (in USD)", fontsize=12)

plt.show()

```

Bar and count plot

```python
#Bar plot
sns.catplot(x="time", y="total_bill",hue="sex", data=data,kind="bar")
plt.title("Average Total Bill by Time and Gender", fontsize=16, fontweight='bold')
plt.xlabel("Time of Day", fontsize=12)
plt.ylabel("Average Total Bill (in USD)", fontsize=12)

# Display the plot
plt.show()
```

```python
#count plot
sns.catplot(x="total_bill",hue="size", data=data,kind="bar",estimator=np.std,orient='v')

plt.title("Standard Deviation of Total Bill by size Status", fontsize=16, fontweight='bold')
plt.ylabel("Size", fontsize=12)
plt.xlabel("Standard Deviation of Total Bill (in USD)", fontsize=12)
plt.show()
```

HeatMap

```python
#data
df=sns.load_dataset('iris')
```

```python
df['species'] = pd.Categorical(df['species']).codes
```

```python
# Plotting the heatmap
plt.figure(figsize=(10, 6)) 
sns.heatmap(
    df[:20], 
    annot=True, cmap="coolwarm",  
    linewidths=0.5, linecolor="gray",  
    cbar_kws={"shrink": 0.8, "label": "Value Range"} 
)

plt.title("Heatmap of the First 20 Rows", fontsize=16, fontweight="bold")
plt.xlabel("Columns", fontsize=12)
plt.ylabel("Index", fontsize=12)

plt.show()
```

```python
# Use clustermap (numeric columns only)
import scipy as scipy
sns.clustermap(
    data=data,
    metric="euclidean",  # Distance metric
    method="ward",       # Linkage method
    cmap="coolwarm",     # Colormap
    standard_scale=1,    # Normalize columns
    figsize=(10, 8)      # Figure size
)

# Show the plot
plt.title("Cluster Map of Iris Dataset", fontsize=16, fontweight='bold')
plt.show()
```

Joint Plots

```python
#Scatter plot
sns.jointplot(
    x="total_bill", 
    y="tip", 
    data=data, 
    kind="scatter", 
    color="blue", 
    height=4, 
    marginal_kws=dict(bins=25, fill=True)
)

plt.suptitle("Total Bill vs Tip", fontsize=16, fontweight='bold', y=1.02)
plt.show()
```

```python
#hex plot
sns.jointplot(
    x="total_bill", 
    y="tip", 
    data=data, 
    kind="hex", 
    color="red", 
    height=4, 
    marginal_kws=dict(bins=25, fill=True)
)

plt.suptitle("Total Bill vs Tip", fontsize=16, fontweight='bold', y=1.02)
plt.show()
```

```python
#hex plot
sns.jointplot(
    x="total_bill", 
    y="tip", 
    data=data, 
    kind="kde",  
    cmap="Blues",  
    height=7,  
    shade=True,  
    fill=True,  
    thresh=0.1,  
    marginal_kws={"fill": True}
)

plt.suptitle("Total Bill vs Tip (Kernel Density Estimation)", fontsize=18, fontweight='bold', y=1.02)
plt.show()
```

```python
#Regrresion plot
sns.jointplot(
    x="total_bill", 
    y="tip", 
    data=data, 
    kind="reg",  
    height=7,  
    
)

plt.suptitle("Total Bill vs Tip (Kernel Density Estimation)", fontsize=18, fontweight='bold', y=1.02)
plt.show()
```

Pair plot

```python
sns.pairplot(data, hue="sex", palette="", markers=["o", "s"], height=2.5)

plt.suptitle("Pairplot of Tips Dataset by Sex", fontsize=16, fontweight='bold', y=1.02)
plt.show()
```
